### Hierarchial Clustering

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('Customers.csv')
# dataset.head()
```

```python
# In Unsupervised we not need any y-var.(dependent var.) to predict for output , whichever column going to consider is 
# into this X var. (all independent var.)

X = dataset.iloc[:,[3,4]].values   # [Annual Income, Spending Score]
```

```python
# Using the Dendogram to determine the no. of clusters 

import scipy.cluster.hierarchy as sch

dendogram = sch.dendrogram(sch.linkage(X, method='ward'))  # creating link of sch

# plt.xticks(rotation=45, size=50)
plt.title('Dendogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean Distance')
plt.show()
```

```python
# Create the model , to fit the hierarchial clustering

from sklearn.cluster import AgglomerativeClustering

hc = AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='ward')

y_hc = hc.fit_predict(X)  # shows how many clusters created 

# affinity: what type of parameter we used internally
```

```python
y_hc   # here we have 5 clusters : 0 to 4
```

#### Plotting the clusters

```python
plt.scatter(X[y_hc==0,0], X[y_hc==0,1], s=100, c='red', label='Cluster1')        # clusters 0 to 4
# scatter(x-var, y-var, s-is pixels)  
# x-var: y_hc= 0th cluster , 0th column of X 
# y-var: y_hc= 0th cluster , 1st cloumn of X   , similar for all clusters

plt.scatter(X[y_hc==1,0], X[y_hc==1,1], s=100, c='blue', label='Cluster2')
plt.scatter(X[y_hc==2,0], X[y_hc==2,1], s=100, c='green', label='Cluster3')
plt.scatter(X[y_hc==3,0], X[y_hc==3,1], s=100, c='cyan', label='Cluster4')
plt.scatter(X[y_hc==4,0], X[y_hc==4,1], s=100, c='magenta', label='Cluster5')

plt.title('Clusters of the Customers')
plt.xlabel('Annual Income(K-$)')
plt.ylabel('Spending Score(1-100)')
plt.legend()
plt.show()
```
